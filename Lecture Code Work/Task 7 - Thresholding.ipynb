```python
# Boiler plate imports 
import numpy as np
import cv2
from matplotlib import pyplot as plt
import easygui
```

```python
# Loading in the imgages
googly = plt.imread("data/googly.jpg")
sudoku = plt.imread("data/sudoku.jpg")

googly = cv2.cvtColor(googly,cv2.COLOR_BGR2GRAY)
sudoko = cv2.cvtColor(sudoku,cv2.COLOR_BGR2GRAY)
```

```python
# Subplot setup
fig = plt.figure(figsize=(20,20))

ax1 = plt.subplot2grid((2,2), (0,0),)
ax2 = plt.subplot2grid((2,2), (0,1),)
ax3 = plt.subplot2grid((2,2), (1,0),)
ax4 = plt.subplot2grid((2,2), (1,1),)

T, B = cv2.threshold(googly, thresh = 150, maxval = 255, type = cv2.THRESH_BINARY)
threshedGoogly = cv2.cvtColor(B, cv2.COLOR_GRAY2BGR)

ax1.imshow(cv2.cvtColor(googly,cv2.COLOR_GRAY2BGR))
ax1.set_title("Original image")

T, B = cv2.threshold(googly, thresh = 50, maxval = 255, type = cv2.THRESH_BINARY)
threshedGoogly = cv2.cvtColor(B, cv2.COLOR_GRAY2BGR)

ax2.imshow(threshedGoogly)
ax2.set_title("Threshold level = " + str(T))

T, B = cv2.threshold(googly, thresh = 100, maxval = 255, type = cv2.THRESH_BINARY)
threshedGoogly = cv2.cvtColor(B, cv2.COLOR_GRAY2BGR)

ax3.imshow(threshedGoogly)
ax3.set_title("Threshold level = " + str(T))

T, B = cv2.threshold(googly, thresh = 150, maxval = 255, type = cv2.THRESH_BINARY)
threshedGoogly = cv2.cvtColor(B, cv2.COLOR_GRAY2BGR)

ax4.imshow(threshedGoogly)
ax4.set_title("Threshold level = " + str(T))
```
