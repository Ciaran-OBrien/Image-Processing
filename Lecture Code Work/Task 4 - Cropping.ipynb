```python
# Boiler plate imports 
import numpy as np
import cv2
from matplotlib import pyplot as plt
from matplotlib import image as image
import easygui
```

```python
# Loading in the imgage
ballPitImage = image.imread("data/ballpit.jpg")
```

```python
# Subplot setup
fig = plt.figure(figsize=(20,20))

ax1 = plt.subplot2grid((2,2), (0,0),)
ax2 = plt.subplot2grid((2,2), (0,1),)
ax3 = plt.subplot2grid((2,2), (1,0),)

# Show the normal version of the image
ax1.imshow(ballPitImage)

# Crop the image
height, width, channels = ballPitImage.shape
widthIndent = int(round(width*0.25))
heightIndent = int(round(height*0.25))
croppedBallPit = ballPitImage[height:(height-heightIndent),width:(width-widthIndent)]
cropped = ballPitImage[100:width-widthIndent,0:height-heightIndent] # y,x
ax2.imshow(cropped)

# Rotate the image
matrixBallPit = cv2.getRotationMatrix2D(center=(width/2,height/8), angle=45, scale=1)
rotatedBallPit = cv2.warpAffine(cropped, M = matrixBallPit, dsize=(width,height))
ax3.imshow(rotatedBallPit)

plt.show()
```
