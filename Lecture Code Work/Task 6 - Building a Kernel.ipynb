```python
# Boiler plate imports 
import numpy as np
import cv2
from matplotlib import pyplot as plt
from matplotlib import image as image
import easygui
```

```python
# Loading in the imgage
ballPitImage = image.imread("data/ballpit.jpg")
```

```python
# Creating a HPF filter to apply to the ball pit image 
kernel = np.array([[-1,-1,-1], [-1,9,-1], [-1,-1,-1]], dtype=float)
filterImage = cv2.filter2D(ballPitImage,ddepth=-1,kernel=kernel)
subtractedImage = ballPitImage - filterImage
```

```python
# Subplot setup
fig = plt.figure(figsize=(20,20))

ax1 = plt.subplot2grid((3,1), (0,0),)
ax2 = plt.subplot2grid((3,1), (1,0),)
ax3 = plt.subplot2grid((3,1), (2,0),)

ax1.imshow(ballPitImage)
ax1.set_title("Original Image")
ax2.imshow(filterImage)
ax2.set_title("Sharpened Image")
ax3.imshow(subtractedImage)
ax3.set_title("Subtracted Image")
plt.show()
```
